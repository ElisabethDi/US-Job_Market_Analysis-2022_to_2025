```python
# Created a function that will handle missing values in Data Frame

def handle_missing_values(df, column, fill_value=None):
    """
    Show missing values and optionally return a column with NaN values filled.
    """
    missing = df[column].isnull().sum()
    print(f"Missing values in '{column}': {missing}")
    
    if fill_value is not None:
        print(f"Filling missing values in '{column}' with: {fill_value}")
        return df[column].fillna(fill_value)
    
    return df[column]
```

```python
# Role Extraction Function

# Inporting library
import re


# Extract job role from title

def extract_role(title):
    if pd.isnull(title):
        return None

    title = title.lower()

    # Check specific patterns to avoid misclassification
    patterns = {
        'Clinical Data Analyst': r'clinical.*data.*analyst',
        'Digital Data Analyst': r'(digital|content).*data.*analyst',
        'Cyber Security Analyst': r'(cyber\s*security|security).*analyst',
        'Financial Analyst': r'financial.*analyst',
        'Fraud Analyst': r'fraud.*analyst',
        'Risk Analyst': r'risk.*analyst',
        'Operations Research Analyst': r'research.*analyst',
        'Procurement Analyst': r'procurement.*analyst',
        'Product Analyst': r'product.*analyst',
        'Supply Chain Analyst': r'supply\s*chain.*analyst',
        'Transportation Analyst': r'transportation.*analyst',
        'Human Resources Analyst': r'(human\s*resources|hr|workforce).*analyst',
        'Healthcare Analyst': r'(healthcare|health).*analyst',
        'Marketing Analyst': r'marketing.*analyst',
        'Business Analyst': r'business.*analyst',
        'Operations Data Analyst': r'operations.*analyst',
        'Data Governance Analyst': r'data\s*governance.*analyst',
        'Data Governance Analyst':r'(steward|stewardship).*analyst',
        'Data Scientist': r'data\s*scientist',
        'Data Engineer': r'data\s*engineer',
        'Data Analyst': r'data.*analyst',
        'Actuarial Analyst': r'actuarial.*analyst',
        'Environmental Analyst': r'environmental.*analyst'
        
    }

    for role, pattern in patterns.items():
        if re.search(pattern, title):
            return role

    return None
```

```python
# Seniority Extraction Function
# Extract seniority level from title

def extract_seniority(title):
    if pd.isnull(title):
        return None

    title = title.lower()

    if re.search(r'\bintern\b', title):
        return 'Intern'
    elif re.search(r'\bjunior\b', title) or re.search(r'\bjr\b', title):
        return 'Junior'
    elif re.search(r'\bmid\b', title) or re.search(r'\bintermediate\b', title):
        return 'Mid-Level'
    elif re.search(r'\bsenior\b', title) or re.search(r'\bsr\b', title):
        return 'Senior'
    elif re.search(r'\blead\b', title):
        return 'Lead'
    elif re.search(r'\bprincipal\b', title):
        return 'Principal'
    elif re.search(r'\bmanager\b', title):
        return 'Manager'
    elif re.search(r'\bdirector\b', title):
        return 'Director'
    elif re.search(r'\bchief\b|\bcdo\b|\bhead\b', title):
        return 'Executive'
    else:
        return ''
```

```python
# Apply to df_copy DataFrame

df_copy['clean_title'] = df_copy['title'].apply(extract_role)

df_copy['seniority_level'] = df_copy['title'].apply(extract_seniority)
```

```python
# Function created to  return clean website column

import re


def extract_website_source(raw_name):
    if pd.isnull(raw_name):
        return None

    raw_name = raw_name.lower().strip()

    patterns = {
        'Job Connect': r'job connection hub',
        'Recruiter.com': r'recruiter(\.com)?|recruiter jobs',
        'Prime Recruitment': r'prime job seekers',
        'aijobs.net': r'(aijobs|ai-jobs|aijobs\.ai|aijobs\.ai|aijobs\.net|ai-jobs\.net|jobs - towards ai|jobs and talents in ai)',
        'HiringsArena': r'hiringsarena',
        'Talent.com': r'talent\.com',
        'Snagajob': r'snagajob',
        'Get.It': r'get\.it',
        'LaborMAX': r'labormax',
        'The Ladders': r'\bladders\b',
        'AAAI Career Center': r'aaai career center',
        'iCrunchData': r'icrunchdata',
        'CBRE': r'cbre jobs?',
        'JMIR Career Center': r'jmir',
        'SimplyHired': r'simplyhired',
        'Talentify': r'talentify',
        'Workday': r'workday',
        'Honeywell Careers': r'honeywell',
        'HERC': r'higher education recruitment consortium|herc',
        'JobServe': r'jobserve',
        'Womenforhire': r'women for hire',
        'Lightning Bolt Scheduling': r'lightningbolt',
        'Bebee': r'trabajo\.org|bebee',
        'Upwork': r'\bupwork\b.*|.*\bupwork\b.*'
    }

    for standard_name, pattern in patterns.items():
        if re.search(pattern, raw_name):
            return standard_name

    return raw_name  # fallback if no match
```

```python
# Apply extract_website_source function to clean website column

df_copy['website'] = df_copy['website'].apply(extract_website_source)
```
